Living is Hard

Life is an unusual state of matter. It is complex, unstable, and requires constant energy to maintain. In fact it always seem to be on the verge of ending and reverting back into non-life.

Imagine you are a single celled creature in the oceans of primordial earth. If you get washed up on shore and dry out you die. If you don't run into nutrients on a regular basis, you die. If your internal chemical state gets thrown off because the water around you is a bit too salty, or not salty enough, you die.

As evolution has progressed it has managed to keep life going, but its never been easy. Every moment is a struggle to maintain the specific chemical and internal environmental states that allow life to continue. This balancing act is called maintaining homeostasis. Homeostasis is the tendency of organisms to auto-regulate and maintain their internal environment in a stable state.

Without homeostasis everything that life needs to be life would quickly spiral out of control. We'd get too hot, too cold, starve to death, or dry out and get blown away as dust. The process of controlling chemicals and states exists at every level of biology and evolution. It's no wonder that homeostasis would be critical for humans, human behavior and intelligence itself.
